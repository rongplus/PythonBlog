```python
workbook = xlsxwriter.Workbook('result.xlsx')
worksheet = workbook.add_worksheet()
index=0
def open_local(url):
    
    dom = etree.HTML(url)
    #return
    data = open(url,'r').read()
    doc = etree.HTML(data)
    
    iiii=0
    

    for tr in doc.xpath('//div[@class="link-item status-sld hasheader"]'):
        #print (tr.xpath('.text()'))
        with open('readme'+ str(iiii)+'.txt', 'w',encoding="utf-8") as f:
            aa = str(etree.tostring(tr))
            f.write(aa)
        iiii +=1
    return
    
    with open(url) as fp:
        soup = BeautifulSoup(fp, 'html.parser')
    
        #page = urlopen(monthly_url)
        #soup = BeautifulSoup(page, 'html.parser')
        lt = soup.find_all('div', attrs={'class': 'link-item status-sld hasheader'})
        print("----link-item status-sld hasheader =", len(lt))
        for ii in lt:
            findInPart(ii)
            
def open_remote(monthly_url):
    page = urlopen(monthly_url)
    soup = BeautifulSoup(page, 'html.parser')
    lt = soup.find_all('div', attrs={'class': 'link-item status-sld hasheader'})
    print("----link-item status-sld hasheader =", len(lt))
    for ii in lt:
        findInPart(ii)
    
def findInPart(part):
    lt = part.find_all('div', attrs={'class': 'formitem formgroup tabular'})
    if (len(lt)==0):
        print("find Error--------------")
        return

    print("----formitem formgroup tabular =", len(lt))
    findInTabular1(lt[0])

def findInTabular(tabular):
    lt = tabular.find_all('div', attrs={'class': 'formitem formgroup horizontal'})
    print("----formitem formgroup horizontal =", len(lt))

def findInHorizontal(horizontal):
    lt = horizontal.find_all('div', attrs={'class': 'formitem formgroup horizontal'})

    
def findInTabular1(tabular):
    global index
    lt = tabular.find_all('span',attrs="formitem formfield")
    print("----in tabular =", len(lt))
    data_list = []
    for span in lt:
        print(span.text,"---")
        data_list.append(span.text)
    writeToExel(index,data_list)
    index+=1
       
def writeToExel(row,data_list):
        for col_num, data in enumerate(data_list):
            worksheet.write(row, col_num, data)
        
url='C:\\Users\\huang\\Downloads\\a.htm'
#open_local(url)
```
